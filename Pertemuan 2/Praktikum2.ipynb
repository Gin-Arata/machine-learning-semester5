```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Penjelasan import sklearn
# LabelEncoder digunakan untuk mengubah variabel nominal menjadi variabel ordinal
# StandardScaler digunakan untuk menstandarisasi variabel numerik sehingga memiliki mean 0 dan standar deviasi 1
```

# Load Data

```python
data = "./data/Titanic-Dataset-fixed.csv"
df = pd.read_csv(data)
df.head()

# Penjelasan Kode Diatas
# pd.read_csv(data) digunakan untuk membaca file csv yang berada di dalam variabel data
# df.head() digunakan untuk menampilkan 5 baris pertama dari data
```

# Slice Data

```python
df = df[['Survived', 'Pclass', 'Sex', 'Age', 'Cabin']]
df.head()

# Penjelasan Kode Diatas
# df[['Survived', 'Pclass', 'Sex', 'Age', 'Cabin']] : Digunakan untuk slicing data yang akan digunakan. Jadi panda hanya akan mengambil data yang ada di dalam list tersebut
```

# Encoding


```python
# Melakukan encoding perlu dilakukan karena machine learning membaca data dan kalkulasi dalam bentuk numerik

le = LabelEncoder() # Membuat objek LabelEncoder
df['Sex'] = le.fit_transform(df['Sex'])
df['Cabin'] = le.fit_transform(df['Cabin'])

# Penjelasan kode diatas
# le.fit_transform(df['Sex/Cabin']) : Digunakan untuk mengubah data nominal menjadi data ordinal. 'Fit' digunakan untuk mempelajari semua nilai unik pada kolom, 'Transform' digunakaan untuk mengubah setiap nilai dalam kolom menjadi nilai ordinal yang sesuai
```

# Verifikasi Hasil

```python
df.head()
```

# Standarisasi

```python
# Melakukan standarisasi terhadap age

std = StandardScaler() # Membuat objek StandardScaler
df['Age'] = std.fit_transform(df[['Age']])

# Penjelasan kode diatas
# fit_transform : 'fit' Digunakan untuk menghitung mean dan standar deviasi, 'transform' Digunakan untuk menstandarisasi nilai nilai dalam kolom sehingga memiliki nilai rerata 0 dan standar deviasi 1
# Penggunaan 2 kurung siku dalam fit_transform karena diharapkan memiliki output DataFrame atau Array 2D dan jika hanya memiliki 1 kurung siku output akan memiliki Array 1D atau Series
```
