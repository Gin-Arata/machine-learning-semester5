```python
import pandas as pd
```

# Load File CSV

```python
data = "Titanic-Dataset.csv"
df = pd.read_csv(data) # Untuk load files csv agar bisa diolah
```

# Menampilkan 5 Data Teratas

```python
# Menampilkan data 5 teratas dari file Titanic-Dataset.csv

df.head() # Untuk menampilkan data 5 teratas dari file
```

# Pengecekan Data Type

```python
# Menampilkan informasi terhadap data

df.info() # Untuk menampilkan informasi terhadap data
```

# Menghitung Jumlah Null pada Datasets

```python
# Untuk menghitung jumlah data null pada dataset

df.isnull().sum() # Function tersebut akan melakukan pengecekan terhadap data yang kosong kemudian akan dijumlahkan setiap data yang null
```

# Imputasi Data

```python
# Melakukan imputasi terhadap data yang null
# imputasi adalah proses pengisian data yang kosong dengan data yang sesuai

# Melakukan imputasi terhadap age
df['Age'] = df['Age'].fillna(value=df['Age'].mean())

# Penjelasan kode diatas
# fillna : Function yang digunakan untuk mengisi data yang kosong
# value : Parameter yang digunakan untuk mengisi data yang kosong
# inplace : Parameter yang digunakan untuk mengganti data yang kosong dengan data yang diisi


# Melakukan imputasi terhadap Cabin
df['Cabin'] = df['Cabin'].fillna(value="DECK")

# Penjelasan kode diatas
# fillna : Function yang digunakan untuk mengisi data yang kosong
# value : Parameter yang digunakan untuk mengisi data yang kosong
# mean() : Function mean digunakan untuk mengitung nilai rata-rata dari data
# inplace : Parameter yang digunakan untuk mengganti data yang kosong dengan data


# Melakukan imputasi terhadap Embarked
df['Embarked'] = df['Embarked'].fillna(value=df['Embarked'].mode()[0])

# Penjelasan kode diatas
# fillna : Function yang digunakan untuk mengisi data yang kosong
# value : Parameter yang digunakan untuk mengisi data yang kosong
# mode()[0] : Function mode digunakan untuk menghitung nilai yang paling sering muncul terhadap dataset [0] digunakan untuk mengambil nilai pertama pada mode
# inplace : Parameter yang digunakan untuk mengganti data yang kosong dengan value yang ditentukan sebelumnya
```
